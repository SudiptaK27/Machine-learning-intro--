Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS).

Answer:

Artificial Intelligence (AI):

-Broad field aiming to build systems that can mimic human intelligence.

-Includes reasoning, decision-making, problem-solving, natural language understanding, etc.

-Example: Chatbots like Siri or Alexa.

Machine Learning (ML):

-Subset of AI that enables systems to learn from data without being explicitly programmed.

-Uses algorithms to find patterns and improve over time.

-Example: Spam email filtering.

Deep Learning (DL):

-Subset of ML based on artificial neural networks with multiple layers.

-Excels at handling large, complex datasets like images, audio, and text.

-Example: Self-driving cars detecting pedestrians using computer vision.

Data Science (DS):

-Broader discipline that uses statistics, ML, and domain knowledge to extract insights from structured and unstructured data.

-Involves data cleaning, visualization, and storytelling.

-Example: Analyzing customer purchase data to suggest product recommendations.



Question 2: What are the types of machine learning? Describe each with one
real-world example.

Answer:

There are three types of machine learning

**Supervised Learning:**

Learns from labeled data (input + output).

Goal: Predict outcomes for unseen inputs.

Example: Predicting house prices using features like size and location.

**Unsupervised Learning:**

Learns from unlabeled data.

Goal: Find patterns, clusters, or hidden structures.

Example: Customer segmentation in marketing.

**Reinforcement Learning (RL):**

Agent interacts with an environment, learns via rewards/punishments.

Goal: Maximize cumulative reward.

Example: Training a robot to walk or AlphaGo winning board games.


Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

Answer:

**Overfitting:**

Model learns training data too well, including noise.

High accuracy on training data, poor performance on test data.

Example: A decision tree with too many branches memorizing the dataset.

**Underfitting:**

Model is too simple to capture patterns.

Poor performance on both training and test data.

Example: A linear model trying to fit non-linear data.

**Bias-Variance Tradeoff:**

Bias: Error from oversimplifying the model (underfitting).

Variance: Error from making the model too complex (overfitting).

Goal: Find a balance where the model generalizes well.

Question 4: What are outliers in a dataset, and list three common techniques for
handling them.

Answer:

**Outliers:**
Outliers are data points that significantly differ from the rest of the dataset. They can occur due to measurement errors, data entry mistakes, or rare events. Outliers can distort statistical measures (mean, variance) and negatively impact machine learning models.

Example: In a dataset of monthly incomes where most values are between ₹30,000–₹60,000, an entry of ₹5,00,000 would be an outlier.

Three common techniques to handle outliers:

**Removal (Deleting Outliers):**

If outliers are due to errors or irrelevant records, they can be removed.

Example: Dropping sensor values recorded due to malfunction.

Transformation:

Apply mathematical functions (log, square root, Box-Cox) to reduce the influence of extreme values.

Example: Applying log transformation on income data to normalize distribution.

**Capping/Clipping (Winsorization):**

Replace extreme values beyond a certain percentile (e.g., 1st and 99th percentile) with boundary values.

Helps retain data while reducing outlier impact.


Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.


Answer:

**Missing values:**
Missing values occur when no data value is stored for a variable. They can arise from human error, system issues, or skipped survey questions. If not handled properly, they can lead to biased analysis and poor model performance.

**Process of handling missing values:**

Identify missing values – using methods like .isnull() or .info() in Python.

Analyze missingness pattern – check if data is Missing Completely at Random (MCAR), Missing at Random (MAR), or Not Missing at Random (NMAR).

**Decide strategy –**

Remove records if missingness is very high.

Impute values if missingness is moderate and data is important.

Apply suitable imputation – based on variable type (numerical/categorical).

Imputation techniques:

**Numerical Data:**

Replace missing values with the mean (or median if skewed).

Example: If some "Age" values are missing, fill them with the average age of all entries.

**Categorical Data:**

Replace missing values with the mode (most frequent category).

Example: If "Gender" has missing entries, fill them with the most common value, say "Male" or "Female".

```python
'''Question 8: Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.
'''
#Answer

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate 1000 samples from normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)
df = pd.DataFrame(data, columns=['Value'])

# Introduce 50 random missing values
missing_indices = np.random.choice(df.index, 50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

# Plot before imputation
plt.hist(df['Value'].dropna(), bins=30)
plt.title("Before Imputation")
plt.show()

# Fill missing values with mean
df['Value'].fillna(df['Value'].mean(), inplace=True)

# Plot after imputation
plt.hist(df['Value'], bins=30)
plt.title("After Imputation")
plt.show()


```

Question 10: You are working as a data scientist for a retail company. You receive a customer
transaction dataset that contains:
● Missing ages,
● Outliers in transaction amount,
● A highly imbalanced target (fraud vs. non-fraud),
● Categorical variables like payment method.


Answer:

When preparing a dataset for machine learning, the goal is to ensure data quality, consistency, and fairness before training a model. The dataset here has missing ages, outliers in transaction amounts, an imbalanced target (fraud vs. non-fraud), and categorical variables (e.g., payment method). The step-by-step preparation plan is as follows:

**1. Handling Missing Ages**

First, check the proportion and pattern of missing values.

If only a few values are missing, imputation is preferred rather than dropping rows.

For numerical data like Age, the median is often used because it is robust to skewed distributions.

Additionally, a binary “missing age” indicator can be created to capture whether the age was missing, since missingness itself may carry information.

**2. Handling Outliers in Transaction Amount**

Outliers are extreme values that may distort analysis and model training.

Detect them using statistical methods such as the Interquartile Range (IQR) or z-score.

If outliers are due to errors, they can be removed.

If they are genuine but extreme values, they can be capped at reasonable limits (winsorization) or transformed using methods such as the logarithmic transformation to reduce skewness.

This ensures the model is not overly influenced by very large transactions.

**3. Handling Class Imbalance (Fraud vs Non-fraud)**

Fraud datasets are usually highly imbalanced, with far fewer fraud cases.

This imbalance can cause the model to ignore the minority (fraud) class.

Possible solutions include:

Resampling: Oversample the minority class (e.g., SMOTE) or undersample the majority class.

Algorithm-level approaches: Use models that allow class weighting so that fraud cases are given more importance.

Evaluation metrics: Instead of accuracy, use metrics like precision, recall, F1-score, or PR-AUC that better reflect minority class performance.

**4. Encoding Categorical Variables (Payment Method, etc.)**

Categorical variables cannot be directly used by most ML algorithms.

If the variable has a small number of categories (e.g., Cash, Credit, Debit, UPI), use One-Hot Encoding.

For variables with many categories, alternatives like frequency encoding or target encoding may be more suitable.

This ensures the categorical information is preserved without introducing bias.

**5. Final Preparation and Pipeline**

After the above steps, all features should be standardized to a consistent format.

Numerical variables may be scaled if required (especially for algorithms sensitive to scale, like logistic regression or SVM).

Keep the data processing consistent by applying the same transformations on both training and test datasets.

Finally, ensure that resampling methods for imbalance are applied only on the training set to avoid data leakage.
